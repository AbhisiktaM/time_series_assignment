Data cleaning for USD

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\user\C drive files\Documents\gold\Monthly_Avg.csv")
```

```python
df['Year']=df['Date'].apply(lambda x: str(x)[-4:])
df['Month']=df['Date'].apply(lambda x: str(x)[:-8])
df['Day']=df['Date'].apply(lambda x: str(x)[-7:-5])

# Changed MM/DD/YY -> YY/MM/DD

df['Dates']=pd.DatetimeIndex(df['Year']+'-'+df['Month']+'-'+df['Day'] )
```

```python
df.drop(['Date','EUR','JPY','GBP','CAD','CHF','CNY','TRY','SAR','IDR','AED','THB','VND','EGP','KRW','RUB','ZAR','AUD','Year','Month','Day'], 
        axis=1, inplace=True, errors='ignore')
```

```python
print(df.columns)
```

```python
df.columns = ['USD', 'INR','Dates']
```

```python
df['USD'] = df['USD'].apply(lambda x: float(str(x).replace(',', '')))
df_usd = df[['Dates', 'USD']].dropna()
```

```python
# Train-test split: last 12 months
train = df_usd['USD'].iloc[:-12]
test = df_usd['USD'].iloc[-12:]
```

```python
# Fit ARIMA model (order can be optimized further)
model_arima = ARIMA(train, order=(5, 1, 0))
fit_arima = model_arima.fit()
```

```python
forecast_arima = fit_arima.forecast(steps=12)
```

```python
mae_arima = mean_absolute_error(test, forecast_arima)
rmse_arima = np.sqrt(mean_squared_error(test, forecast_arima))
```

```python
plt.figure(figsize=(14, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test', color='green')
plt.plot(test.index, forecast_arima, label='ARIMA Forecast', color='purple', linestyle='--')
plt.title('Gold Price Forecast using ARIMA')
plt.xlabel('Date')
plt.ylabel('USD Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

mae_arima, rmse_arima
```

Outlier Detection for ARIMA using z-scores (USD)

```python
df_usd['z_score'] = (df_usd['USD'] - df_usd['USD'].mean()) / df_usd['USD'].std()
outliers_a_usd = df_usd[np.abs(df_usd['z_score']) > 3]
# Plot the time series with outliers marked
plt.figure(figsize=(14, 6))
plt.plot(df_usd['Dates'], df_usd['USD'], label='Gold Price (USD)')
plt.scatter(outliers_a_usd['Dates'], outliers_a_usd['USD'], color='red', label='Outliers', zorder=5)
plt.title('Outlier Detection (Z-score > 3)')
plt.xlabel('Date')
plt.ylabel('USD Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Outlier Detection using ARIMA specific detection (USD)

```python
import numpy as np
import pandas as pd
import pmdarima as pm

# Fit the ARIMA model with outlier detection
model = pm.auto_arima(df_usd['USD'], seasonal=True, m=12, stepwise=True, trace=True, out_of_sample_size=12)

# Print outlier detection results
print(model.summary())

# Forecast and plot results
forecast, conf_int = model.predict(n_periods=12, return_conf_int=True)

plt.figure(figsize=(14, 6))
plt.plot(df_usd['USD'], df_usd['USD'], label='Gold Price (USD)')
plt.plot(pd.date_range(df_usd['Dates'].iloc[-1], periods=12, freq='M'), forecast, label='ARIMA Forecast', color='red', linestyle='--')
plt.fill_between(pd.date_range(df_usd['Dates'].iloc[-1], periods=12, freq='M'), conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.3)
plt.title('ARIMA Forecast with Outlier Detection')
plt.xlabel('Date')
plt.ylabel('USD Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
